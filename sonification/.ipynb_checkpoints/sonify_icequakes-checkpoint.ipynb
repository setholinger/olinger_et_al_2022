```python
'''

Define a few useful functions for sonifiying individual events and the entire catalog

'''

# define function to plot timeseries and spectrogram
def plot_event_and_spectra(st,win_len,freq,fs):
    # plot data
    fig,ax = plt.subplots(2,1,figsize=(10,10))
    if type(st) == obspy.core.stream.Stream:
        times = [t.datetime for t in st[0].times("UTCDateTime")]
        ax[0].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        ax[0].plot(times,st[0].data*1000,'k')
    else:
        times = np.arange(0,len(st)/fs,1/fs)
        ax[0].plot(times,st*1000,'k')
    ticks = [times[len(times)//8],times[len(times)//4],times[len(times)//8*3],times[len(times)//2],times[len(times)//8*5],times[len(times)//4*3],times[len(times)//8*7],times[len(times)-1]]
    ax[0].grid(True)
    ax[0].set_ylabel("Velocity (mm/s)")    
    ax[0].set_title("Time domain signal")

    # plot spectrogram 
    if type(st) == obspy.core.stream.Stream:
        f,t,s = spectrogram(st[0].data*1000, fs=fs, nperseg=win_len*fs)
        times = [st[0].stats.starttime.datetime + datetime.timedelta(seconds=time) for time in t]
    else:
        f,t,s = spectrogram(st*1000, fs=fs, nperseg=int(win_len*fs))
        times = t
    spec = ax[1].pcolor(times, np.log10(f), np.log10(s), vmin=np.log10(10e-11),vmax=np.log10(10e-5))
    ax[1].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax[1].get_shared_x_axes().join(ax[0],ax[1])
    ax[1].set_xticks(ticks)
    ax[1].set_xlim([times[0],times[-1]])
    ax[1].set_yticks([np.log10(freq[0]),0,np.log10(freq[1])])
    ax[1].set_ylim([np.log10(f[1]),np.log10(f[-1])])
    ax[1].set_yticklabels(["$10^{-1}$","$10^{0}$","$10^{1}$"])
    ax[1].set_ylabel("Frequency (Hz)")
    ax[1].set_xlabel("Time")
    plt.subplots_adjust(right=0.85)
    cbar_ax = fig.add_axes([0.9, 0.1225, 0.025, 0.5])
    cbar = plt.colorbar(spec,label="PSD [(mm/s)$^2$/Hz]",ticks=[-10,-8,-6,-4],cax=cbar_ax)
    cbar.ax.set_yticklabels(['$10^{-8}$','$10^{-8}$','$10^{-6}$','$10^{-4}$'])
    plt.show()
```

```python
def save_wav(fname,data,fs):
    w = wav.open(fname, 'wb')
    w.setparams((1, 4, fs, len(data), 'NONE','not compressed'))
    maxint = 2 ** (3.75 * 8 - 1) - 1
    data = data / abs(data).max() * maxint
    data = np.require(data, dtype='<i4')
    w.writeframes(data.tobytes())
    return

def sonify(data,fs_in,fs_out,stretch,shift,osc_range,freq_lims,num_bands,filt_bank_type="linear",osc_bank_type="linear",amplification_factor=1,plot=False):

    # make sure fs_out is above the nyquist of the highest audio frequency you want to play!
    
    # get filter bank
    if filt_bank_type == "linear":
        filt_bank = np.linspace(freq_lims[0],freq_lims[1],num_bands+1)
    elif filt_bank_type == "log":
        filt_bank = freq_lims[0] + np.logspace(0,np.log10(100),num_bands+1)/100*freq_lims[1]
    #filt_bank = np.logspace(np.log10(freq_lims[0]),np.log10(freq_lims[1]),num_bands+1)
    #filt_bank = np.flipud(1.01*freq_lims[1] - np.logspace(np.log10(freq_lims[0]),np.log10(freq_lims[1]),num_bands+1))
    
    # get shifted oscillator frequencies
    if osc_bank_type == "linear":
        osc_bank = shift + np.linspace(0,osc_range,num_bands)
    if osc_bank_type == "log":    
        osc_bank = shift + np.logspace(0,np.log10(100),num_bands)/100*osc_range
    #osc_bank = shift + np.flipud(osc_range - np.logspace(0,np.log10(osc_range),num_bands))
    if osc_bank_type == "sigmoid":
        osc_bank = shift + osc_range/(1+np.exp(-np.linspace(-num_bands,num_bands,num_bands)))
    osc_bank *= stretch


    
    # plot frequency mapping function
    if plot:
        envs = []
        fig,ax = plt.subplots(5,1,figsize=(20,20))
        ax[0].set_title("Mapping function ($f_{signal}$ to $f_{osc}$)")
        ax[0].set_xlabel("$f_{signal}$")
        ax[0].set_ylabel("$f_{osc}$")
        ax[0].plot(filt_bank[:-1],osc_bank/stretch)

    # get output signal length
    len_out = len(data)*fs_out/fs_in
        
    # normalize data
    data = data/abs(data).max()
    
    # taper signal to avoid edge effects
    win = tukey(len(data),alpha=0.1)
    data = win*data
    
    # make zero array for reconstructed signal
    osc_sum = np.zeros(int(len_out))
    for f in range(len(filt_bank)-1):

        # design bandpass filter and apply to data
        cutoff = np.array([filt_bank[f],filt_bank[f+1]])
        b, a = butter(2, cutoff/(fs_in/2), 'bandpass')
        filt_data = filtfilt(b, a, data)
        
        # get envelope of signal
        env = np.abs(hilbert(filt_data))
                
        # taper envelope to avoid edge effects
        env = win*env
        
        # resample envelope
        env = resample(env,int(len_out))
        env = np.abs(env)
        
        # accentuate difference in amplitude between envelopes
        env = env**amplification_factor
        
        # save envelope to make plot
        if plot:
            envs.append(env)
        
        # make oscillator at mapped frequency
        t = np.arange(0,len(env)/fs_out,1/(fs_out))
        osc = np.sin(2*np.pi*t*osc_bank[f])
    
        # multiply by envelope
        osc_sum = osc_sum + osc*env
    
    # stretch signal to desired length
    n_samples = int(stretch * len(osc_sum))
    osc_sum = resample(osc_sum,n_samples)

     # plot envelopes 
    if plot:
        env_max = []
        env_max_idx = []
        t = np.arange(0,len(envs[0])/fs_out,1/fs_out)
        for f,env in enumerate(envs):
            ax[1].plot(t,env,label=filt_bank[f])
            env_max.append(np.max(env))
            env_max_idx.append(np.argmax(env))
        ax[1].set_title("Signal envelopes")
        ax[1].set_xlabel("Time (s)")
        ax[1].set_ylabel("Envelope amplitude")
        ax[1].legend()

        # plot envelope amplitude as a function of frequency (for tuning mapping function)
        ax[2].scatter(filt_bank[:-1],env_max)
        ax[2].set_title("Envelope amplitude with frequency")
        ax[2].set_xlabel("Frequency (Hz)")
        ax[2].set_ylabel("Envelope amplitude")

        # plot envelope frequency as a function of time (for tuning mapping function)
        ax[3].scatter(t[env_max_idx],filt_bank[:-1])
        ax[3].set_title("Envelope frequency through time")
        ax[3].set_xlabel("Time (s)")
        ax[3].set_ylabel("$f_{signal}$ (Hz)")

        # plot oscillator frequency as a function of time (for tuning mapping function)
        ax[4].scatter(t[env_max_idx],osc_bank/stretch)
        ax[4].set_title("Oscillator frequency through time")
        ax[4].set_xlabel("Time (s)")
        ax[4].set_ylabel("$f_{osc}$ (Hz)")
    
        # display plot
        plt.tight_layout()
        plt.show()
    
    return osc_sum/abs(osc_sum).max()
```

```python
# define function to pad with zeros to space events properly in time
def event_audio_tracks(detection_times,audio,gap_fs,starttime,endtime):
    bool_ind = np.logical_and(detection_times > starttime,detection_times < endtime)
    detection_times = detection_times[bool_ind]
    tracks = []
    for i in range(len(detection_times)):
        pre_gap = detection_times[i]-starttime
        pre_gap_samples = int(np.round(pre_gap.total_seconds()*gap_fs))
        post_gap = endtime-detection_times[i]
        post_gap_samples = int(np.round(post_gap.total_seconds()*gap_fs))
        track = np.concatenate((np.zeros(pre_gap_samples),audio,np.zeros(post_gap_samples)))
        tracks.append(track)
    return tracks

# function to make pydub audio segment object
def make_audio_segment(data,fs_out,volume):
    maxint = 2 ** (3.75 * 8 - 1) - 1
    data = data / abs(data).max() * volume * maxint 
    data = np.require(data, dtype='<i4')
    audio_segment = AudioSegment(data.tobytes(),frame_rate=fs_out,sample_width=4, channels=1)
    return audio_segment

# function to overlay a bunch of event tracks
def overlay_tracklist(tracklist,fs_out,volumes):
    overlay_track = make_audio_segment(tracklist[0],fs_out,volumes[0])
    for i,track in enumerate(tracklist[1:]):
        overlay_track = overlay_track.overlay(make_audio_segment(track,fs_out,volumes[i]))
    return overlay_track
```

```python
# set timeseries parameters
starttime = datetime.datetime(2013,9,26,0,0)
endtime = datetime.datetime(2013,12,1,0,0)

# get list of tracks for each event with zeros on both ends for timing
gap_fs = 0.1
rift_tracks = event_audio_tracks(detection_times[rift_indices],rift_audio,gap_fs,starttime,endtime)
rift_margin_tracks = event_audio_tracks(detection_times[rift_margin_indices],rift_margin_audio,gap_fs,starttime,endtime)
margin_tracks = event_audio_tracks(detection_times[margin_indices],margin_audio,gap_fs,starttime,endtime)

# get relative amplitude vector to scale volume
region_indices = [margin_indices,rift_indices,rift_margin_indices]
region_volumes = []
for k,indices in enumerate(region_indices):
    bool_ind = [np.logical_and(detection_times[indices] > starttime,detection_times[indices] < endtime)]
    region_waveforms = waveforms[indices]
    amplitudes = np.max(np.abs(region_waveforms[bool_ind]),axis=1)
    volumes = amplitudes/abs(amplitudes).max()*2
    region_volumes.append(volumes)
    
# # scale volume for each event
# margin_tracks = region_volumes[0].reshape(len(region_volumes[0]),1)*margin_tracks
# rift_tracks = region_volumes[1].reshape(len(region_volumes[1]),1)*rift_tracks
# rift_margin_tracks = region_volumes[2].reshape(len(region_volumes[2]),1)*rift_margin_tracks
 
# combine into a single audio segment for each spatial group
rift_overlay = overlay_tracklist(rift_tracks,fs_out,region_volumes[1])
rift_margin_overlay = overlay_tracklist(rift_margin_tracks,fs_out,region_volumes[2])
margin_overlay = overlay_tracklist(margin_tracks,fs_out,region_volumes[0])

# calculate median backazimuth for each group
rift_tip_backazimuth = np.median(backazimuths[rift_indices])
margin_backazimuth = np.median(backazimuths[margin_indices])
rift_margin_backazimuth = np.median(backazimuths[rift_margin_indices])

# calculate the pan
ice_front_backazimuth = 300
rift_pan = (rift_tip_backazimuth - ice_front_backazimuth)/90
margin_pan = (60 + margin_backazimuth)/90
rift_margin_pan = (rift_margin_backazimuth - ice_front_backazimuth)/90

# pan audio based on mean backazimuth, assuming the listener is facing from the array towards the ice front
rift_panned = rift_overlay.pan(rift_pan)
margin_panned = margin_overlay.pan(margin_pan)
rift_margin_panned = rift_margin_overlay.pan(rift_margin_pan)

# increase volume of each track
db_increase = 0
rift_panned += db_increase
margin_panned += db_increase
rift_margin_panned += db_increase

# combine audio files
stereo_icequakes = rift_margin_panned.overlay(rift_panned.overlay(margin_panned))
file_handle = stereo_icequakes.export("../outputs/audio/stereo_icequakes.wav", format="wav")
```

```python
def plot_backazimuths(backazimuths,landsat_scene,array_centroid,station_grid_coords,axes_coords,color_bounds,colors):

    # open LANDSAT imagery file and plot as it is, in EPSG:3031
    sat_data, bounds = read_landsat(landsat_scene)

    # Construct figure and axis to plot on
    fig,ax = plt.subplots(figsize=(15,15))
    fig.patch.set_facecolor('black')
    ax_image = fig.add_axes(axes_coords)
    ax.axis('off')

    # get corners of imagery extent
    p2 = Proj("EPSG:3245",preserve_units=False)
    p1 = Proj(proj='latlong',preserve_units=False)

    # plot imagery
    horz_len = bounds[2]-bounds[0]
    vert_len = bounds[3]-bounds[1]
    plot_bounds = np.array([bounds[0]+0.35*horz_len,bounds[2]-0.25*horz_len,bounds[1]+0.25*vert_len,bounds[3]-0.4*vert_len])
    ax_image.imshow(sat_data,cmap='gray',extent=[bounds[0],bounds[2],bounds[1],bounds[3]])    

    # define, transform, and plot lat/lon grid
    lat = [-74,-74.5,-75,-75.5]
    lon = [-98,-100,-102,-104]
    x_lab_pos=[]
    y_lab_pos=[]
    line = np.linspace(-110,-90,100)
    for i in lat:
        line_x,line_y = transform(p1,p2,line,np.linspace(i,i,100))
        ax_image.plot(line_x,line_y,linestyle='--',linewidth=2,c='k',alpha=0.25)
        y_lab_pos.append(line_y[np.argmin(np.abs(line_x-plot_bounds[0]))])
    line = np.linspace(-80,-70,100)
    for i in lon:
        line_x,line_y = transform(p1,p2,np.linspace(i,i,100),line)
        ax_image.plot(line_x,line_y,linestyle='--',linewidth=2,c='k',alpha=0.25)
        x_lab_pos.append(line_x[np.argmin(np.abs(line_y-plot_bounds[2]))])

    # set ticks and labels for lat/lon grid
    ax_image.set_xticks(x_lab_pos)
    ax_image.set_xticklabels(labels=[str(lon[0]) + '$^\circ$',str(lon[1]) + '$^\circ$',str(lon[2]) + '$^\circ$',str(lon[3]) + '$^\circ$'],fontsize=20,color='w')
    ax_image.set_xlabel("Longitude",fontsize=20,color='w')
    ax_image.set_yticks(y_lab_pos)
    ax_image.set_yticklabels(labels=[str(lat[0]) + '$^\circ$',str(lat[1]) + '$^\circ$',str(lat[2]) + '$^\circ$',str(lat[3]) + '$^\circ$'],fontsize=20,color='w')
    ax_image.set_ylabel("Latitude",fontsize=20,color='w')
    ax_image.yaxis.set_label_coords(-0.03,0.5)
    ax_image.xaxis.set_label_coords(0.5,-0.03)
    ax_image.set_xlim([plot_bounds[0],plot_bounds[1]])
    ax_image.set_ylim([plot_bounds[2],plot_bounds[3]])
    #ax_image.set_title("Event Backazimuths",fontsize=25)

    # plot station locations
    ax_stats = fig.add_axes(axes_coords)
    ax_stats.scatter(station_grid_coords[:,0],station_grid_coords[:,1],marker="^",c='black',s=100)
    ax_stats.set_xlim([plot_bounds[0],plot_bounds[1]])
    ax_stats.set_ylim([plot_bounds[2],plot_bounds[3]])
    ax_stats.axis('off')
    
    # plot grounding line
    grounding_line_file = "../data/shapefiles/ASAID_GroundingLine_Continent.shp"
    grounding_lines = gpd.read_file(grounding_line_file)
    pig_mask = geometry.Polygon([(plot_bounds[0],plot_bounds[2]),
                        (plot_bounds[0],plot_bounds[3]),
                        (plot_bounds[1],plot_bounds[3]),
                        (plot_bounds[1],plot_bounds[2]),
                        (plot_bounds[0],plot_bounds[2])])
    pig_gdf = gpd.GeoDataFrame(geometry=[pig_mask],crs="EPSG:3245")
    pig_gdf = pig_gdf.to_crs(grounding_lines.crs)
    pig_grounding_line = grounding_lines.clip(pig_gdf)
    pig_grounding_line=pig_grounding_line.to_crs("EPSG:3245")
    pig_grounding_line.plot(linestyle='--',color='r',ax=ax_image)

    # add legend for spatial groups
    x,y = transform(p1,p2,-102.85,-75.185)
    box = matplotlib.patches.Rectangle((x-1000, y-7000), 20200, 10000, linewidth=1, edgecolor='k', facecolor='w',alpha=0.75)
    ax_stats.add_patch(box)
    ax_stats.text(x, y, "Rift tip",c="#d95f02",fontsize=20)
    ax_stats.text(x, y-3000, "Rift/margin",c="#1b9e77",fontsize=20)
    ax_stats.text(x, y-6000, "Shear margin",c="#7570b3",fontsize=20)
   
    # make box for date
    x,y = (plot_bounds[0]+(plot_bounds[1]-plot_bounds[0])/2-8500,plot_bounds[3]-14500)
    box = matplotlib.patches.Rectangle((x, y), 18500, 4500, linewidth=1, edgecolor='k', facecolor='w',alpha=0.75)
    ax_stats.add_patch(box)

    # add North arrow
    line_x,line_y = transform(p1,p2,np.linspace(-102.6,-102.6,100),np.linspace(-74.65,-74.6,100))
    ax_stats.plot(line_x,line_y,color='w',linewidth = 5)
    ax_stats.scatter(line_x[-1],line_y[-1],marker=(3,0,4),c='w',s=400)
    ax_stats.text(line_x[-1]-2750,line_y[-1]-3750,"N",color='w',fontsize=20)
    
    # add scale bar
    ax_stats.plot([plot_bounds[0]+(plot_bounds[1]-plot_bounds[0])/2-10000,plot_bounds[0]+(plot_bounds[1]-plot_bounds[0])/2+10000],[plot_bounds[2]+10950,plot_bounds[2]+10950],color='k',linewidth = 5)
    ax_stats.text(plot_bounds[0]+(plot_bounds[1]-plot_bounds[0])/2-3500,plot_bounds[2]+12000,"20 km",color='k',fontsize=20)

    # add inset figure of antarctica
    world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
    inset_coords = [axes_coords[0] + axes_coords[2]*0.825, axes_coords[1] + axes_coords[3]*0.76,0.2*axes_coords[2],0.2*axes_coords[3]]
    ax_inset = fig.add_axes(inset_coords,projection = ccrs.SouthPolarStereo())
    ax_inset.set_extent([-180, 180, -90, -65], crs=ccrs.PlateCarree())
    geom = geometry.box(minx=-103,maxx=-99,miny=-75.5,maxy=-74.5)
    ax_inset.add_geometries([geom], crs=ccrs.PlateCarree(), edgecolor='r',facecolor='none', linewidth=1)
    ax_inset.add_feature(cartopy.feature.OCEAN, facecolor='#A8C5DD', edgecolor='none')
    
    return fig,ax_image,ax_stats,plot_bounds
    
def transform_imagery(file,dst_crs):
    with rasterio.open(file) as src:
        transform, width, height = calculate_default_transform(
            src.crs, dst_crs, src.width, src.height, *src.bounds)
        kwargs = src.meta.copy()
        kwargs.update({
            'crs': dst_crs,
            'transform': transform,
            'width': width,
            'height': height
        })

        with rasterio.open(file.split('.TIF')[0] + '_epsg3245.TIF', 'w', **kwargs) as dst:
            for i in range(1, src.count + 1):
                reproject(
                    source=rasterio.band(src, i),
                    destination=rasterio.band(dst, i),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=transform,
                    dst_crs=dst_crs,
                    resampling=Resampling.nearest)
            
def get_station_coordinates(xml_path):
    stat_coords = []
    inv = obspy.read_inventory(xml_path + "/*")
    for s in inv.get_contents()['stations']:
        channel = inv.get_contents()['networks'][0] + "." + s.split(' ')[0].split('.')[1] + "..HHZ"
        lat = inv.get_coordinates(channel)["latitude"]
        lon = inv.get_coordinates(channel)["longitude"]
        stat_coords.append([lon,lat])
    _, idx = np.unique(stat_coords,axis=0,return_index=True)
    stat_coords = np.array(stat_coords)[np.sort(idx)]
    return stat_coords

def get_station_grid_locations(station_lon_lat_coords,crs):
    # convert station coordinates to x and y and take average station location
    p2 = Proj(crs,preserve_units=False)
    p1 = Proj(proj='latlong',preserve_units=False)
    [stat_x,stat_y] = transform(p1,p2,station_lon_lat_coords[:,0],station_lon_lat_coords[:,1])
    return np.stack((stat_x,stat_y),axis=1)   

def read_landsat(landsat_scene):
   # read each band
    landsat_B2 = rasterio.open(landsat_scene.split("_B2_")[0]+'_B2_epsg3245.TIF') #blue
    landsat_B3 = rasterio.open(landsat_scene.split("_B2_")[0]+'_B3_epsg3245.TIF') #green
    landsat_B4 = rasterio.open(landsat_scene.split("_B2_")[0]+'_B4_epsg3245.TIF') #red
    image_B2 = landsat_B2.read(1)
    image_B3 = landsat_B3.read(1)
    image_B4 = landsat_B4.read(1)

    # crop each band to 99th percentile of brightness
#     image_B2[image_B2 > np.percentile(image_B2,99)] = np.percentile(image_B2,99)
#     image_B3[image_B3 > np.percentile(image_B3,99)] = np.percentile(image_B3,99)
#     image_B4[image_B4 > np.percentile(image_B4,99)] = np.percentile(image_B4,99)

    # combine bands into natural color image
    image_rgb = np.array([image_B2, image_B3, image_B4]).transpose(1,2,0)
    sat_data = (image_rgb * (255 / np.max(image_rgb))).astype(np.uint8)
    return sat_data,landsat_B2.bounds
```

```python
# get indexing info for each region
region_indices = [margin_indices,rift_indices,rift_margin_indices]
region_tracks = [margin_tracks,rift_tracks,rift_margin_tracks]

# get backazimuths for each region and prepare for animation
alpha_steps = 9
peak_alpha = 0.9
decimate = 1000
dims = (len(region_indices),len(region_tracks[0][0]))
backazimuth_start_frames = np.zeros(dims)
backazimuth_frames = np.zeros(dims)
backazimuth_frames[:] = np.nan
decimated_dims = (len(region_indices),len(backazimuth_frames[0,::decimate]))
backazimuth_frames_decimated = np.zeros(decimated_dims)
backazimuth_frames_decimated[:] = np.nan
alphas = np.zeros(decimated_dims)
for k,indices in enumerate(region_indices):

    # get backazimuths in date range for current region
    bool_ind = np.logical_and(detection_times[indices] > starttime,detection_times[indices] < endtime)
    region_backazimuths = backazimuths[indices][bool_ind]

    # find frames where there is an event
    frames = [np.argmax(track > 0) for track in region_tracks[k]]
    backazimuth_start_frames[k,frames] = region_backazimuths

    # fill the following frames with the same backazimuth and set decaying alpha values
    nonzeros = np.nonzero(backazimuth_start_frames[k,:])[0]
    nonzeros = np.concatenate((nonzeros,[len(backazimuth_start_frames[k])]))
    for i in range(len(nonzeros)-1):
        backazimuth_frames[k,nonzeros[i]:nonzeros[i+1]] = backazimuth_start_frames[k,nonzeros[i]]
    backazimuth_frames_decimated[k,:] = backazimuth_frames[k,::decimate]

    nonzeros = np.nonzero(np.diff(backazimuth_frames_decimated[k,:]))[0]+1
    for i in range(len(nonzeros)):
        for j in range(alpha_steps):
            if nonzeros[i]+j > len(alphas[k,:])-1:
                continue
            else:
                alphas[k,nonzeros[i]+j] = peak_alpha - j/alpha_steps
    alphas[k,nonzeros] = peak_alpha
    
# get array of dates for each sample
duration = endtime - starttime
frame_dates = [starttime + f*duration/backazimuth_frames_decimated.shape[1] for f in range(alphas.shape[1])]
```

```python
# set bin colors and backazimuthal ranges corresponding to each color- order needs to be the same as the backazimuth frames and alphas vector!
color_bounds = [[0,180],[180,280],[280,361]]
colors = ["#7570b3","#d95f02","#1b9e77"]

# get station locations and array centroids
station_lon_lat_coords = get_station_coordinates("../data/XML/")
station_grid_coords = get_station_grid_locations(station_lon_lat_coords,"epsg:3245")
array_centroid = np.mean(station_grid_coords,axis=0)

# make list of landsat imagery scenes to use in animation
scenes = glob.glob("../data/imagery/animation/*/*B2_epsg3245.TIF")
scene_dates = [datetime.datetime.strptime(scene.split("/")[-1].split("_")[3],"%Y%m%d") for scene in scenes]
scenes = np.array(scenes)[np.argsort(scene_dates)]
scene_dates = np.sort(scene_dates)

# get background plot
axes_coords = np.array([0.1, 0.025, 0.8, 0.8])
fig,ax_image,ax_stats,plot_bounds = plot_backazimuths(backazimuths,scenes[0],array_centroid,station_grid_coords,axes_coords,color_bounds,colors)

# plot the date
datestring = datetime.datetime.strftime(frame_dates[0],"%Y-%m-%d")
date_text = ax_stats.text(plot_bounds[0]+(plot_bounds[1]-plot_bounds[0])/2-7200,plot_bounds[3]-13200,datestring,color='k',fontsize=20)

# make polar plot centered at array centroid
x_pos = axes_coords[0] + axes_coords[2]*(array_centroid[0]-plot_bounds[0])/(plot_bounds[1]-plot_bounds[0])
y_pos = axes_coords[1] + axes_coords[3]*(array_centroid[1]-plot_bounds[2])/(plot_bounds[3]-plot_bounds[2])
width = 0.3 * axes_coords[3]

# set the parameters of the polar plot
ax_polar = fig.add_axes([x_pos-width/2,y_pos-width/2,width,width], projection = 'polar')
bins = np.linspace(0,2*np.pi,37)
ax_polar.set_theta_zero_location('N')
ax_polar.set_theta_direction(-1)
ax_polar.set_yticks([])
ax_polar.axis('off')
ax_polar.set_ylim(0, 1)
ax_polar.set_rorigin(0)
barcollection = ax_polar.bar(bins[:-1],np.zeros(36),align='edge',width=np.diff(bins),fill='True',edgecolor='black')

# function to update radial histogram
def animate(i):

    # update LANDSAT imagery when date for current frame passes date for each image
    if i == 0:
        global image_num
        image_num = 1
    if frame_dates[i] < scene_dates[image_num] and frame_dates[i+1] > scene_dates[image_num]:
        scene = scenes[image_num]
        sat_data,bounds = read_landsat(scene)
        ax_image.imshow(sat_data,cmap='gray',extent=[bounds[0],bounds[2],bounds[1],bounds[3]])   
        if image_num < len(scene_dates)-1:
            image_num += 1
            
    # print the date
    datestring = datetime.datetime.strftime(frame_dates[i],"%Y-%m-%d")
    date_text.set_text(datestring)
        
    # plot rays
    radius,bins = np.histogram(backazimuth_frames_decimated[:,i]*np.pi/180,bins=np.linspace(0,2*np.pi,37))
    for j, b in enumerate(barcollection):
        b.set_height(radius[j])
        for c in range(len(color_bounds)):
            if bins[j]*180/np.pi >= color_bounds[c][0] and bins[j]*180/np.pi < color_bounds[c][1]:
                b.set_facecolor(colors[c])
                b.set_alpha(alphas[c,i])

# produce the animation
anim=animation.FuncAnimation(fig,animate,repeat=False,blit=False,frames=backazimuth_frames_decimated.shape[1]-1,interval=100)
#anim=animation.FuncAnimation(fig,animate,repeat=False,blit=False,frames=10,interval=100)

# save animation
fps = fs_out/decimate
anim.save('../outputs/video/baz_animation.mp4',writer=animation.FFMpegWriter(fps=fps,extra_args=['-s','3000x3000','-vcodec', 'libx264']),dpi=200)
plt.show()
```

```python
'''

Experiment with ways to manually shift the frequency of a signal

'''

# naively try to shift frequencies for a cosine
fs = np.pi/16
wave = np.cos(np.arange(0,8*np.pi,np.pi/16))
t = np.arange(0,len(wave)*1/fs,1/fs)
wave_fft = fft(wave)
f = fftfreq(len(wave),1/fs)
wave_fft_pos = wave_fft[len(wave_fft)//2:]
wave_fft_neg = wave_fft[:len(wave_fft)//2]
wave_fft_shift_down = np.concatenate((wave_fft_neg,np.zeros(100),wave_fft_pos))
wave_ifft_shift_down = ifft(wave_fft_shift_down)
t_down = np.arange(0,len(wave_ifft_shift_down)*1/fs-1,1/fs)
wave_fft_shift_up = np.concatenate((wave_fft_neg[:-10],wave_fft_pos[10:]))
wave_ifft_shift_up = ifft(wave_fft_shift_up)
t_up = np.arange(0,len(wave_ifft_shift_up)*1/fs,1/fs)

# plot result
plt.figure(figsize=(20,10))
plt.plot(t,wave,label='Original signal')
plt.plot(t_down,wave_ifft_shift_down,label='Frequency shifted down')
plt.plot(t_up,wave_ifft_shift_up,label='Frequency shifted up')
plt.legend()
plt.show()

# try the naive method with real data
fs = 2.1
wave = margin_waveforms[1]
t = np.arange(0,len(wave)*1/fs,1/fs)
wave_fft = fft(wave)
f = fftfreq(len(wave),1/fs)
wave_fft_pos = wave_fft[len(wave_fft)//2:]
wave_fft_neg = wave_fft[:len(wave_fft)//2]
wave_fft_shift_up = np.concatenate((wave_fft_neg[:-500],wave_fft_pos[500:]))
f_up = fftfreq(len(wave_fft_shift_up),1/fs)
wave_ifft_shift_up = ifft(wave_fft_shift_up)
t_up = np.arange(0,len(wave_ifft_shift_up)*1/fs,1/fs)

# plot result- note that this is a bad way to shift frequencies up, because it destroys part of the spectra and causes signal degradation at large shifts
fig,ax = plt.subplots(2,1,figsize=(20,10))
ax[0].plot(f[:len(f)//2+1],np.abs(wave_fft[:len(wave_fft)//2+1]),label='Original signal')
ax[0].plot(f_up[:len(f_up)//2+1],np.abs(wave_fft_shift_up[:len(wave_fft_shift_up)//2+1]),label='Frequency shifted up')
ax[0].legend()
ax[1].plot(t,wave,label='Original signal')
ax[1].plot(t_up,wave_ifft_shift_up,label='Frequency shifted up')
ax[1].legend()
plt.show()

# try a more intelligent method for a cosine
fs = 1/(np.pi/16)
wave = np.cos(np.arange(0,8*np.pi,np.pi/16))
t = np.arange(0,len(wave)*1/fs,1/fs)
wave_fft = fft(wave)
f = fftfreq(len(wave),1/fs)
wave_up = wave * np.cos(np.arange(0,8*np.pi,np.pi/16)*100)
wave_fft_shift_up = fft(wave_up)
f_up = fftfreq(len(wave_fft_shift_up),1/fs)
wave_fft_shift_up[np.logical_and(f_up > 0.01,f_up < 0.02)] = 0
wave_fft_shift_up[np.logical_and(f_up < -0.01,f_up > -0.02)] = 0
wave_ifft_shift_up = ifft(wave_fft_shift_up)
t_up = np.arange(0,len(wave_ifft_shift_up)*1/fs,1/fs)

# plot result- note that we quickly run up against the nyquist frequency, so this isn't a good method either...
fig,ax = plt.subplots(2,1,figsize=(20,10))
ax[0].plot(f[:len(f)//2],np.abs(wave_fft[:len(wave_fft)//2]),label='Original signal')
ax[0].plot(f_up[:len(f)//2],np.abs(wave_fft_shift_up[:len(wave_fft_shift_up)//2]),label='Frequency shifted up')
ax[0].legend()
ax[1].plot(t,wave,label='Original signal')
ax[1].plot(t_up,wave_ifft_shift_up,label='Frequency shifted up')
ax[1].legend()
plt.show()
```
